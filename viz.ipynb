```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
defaultMap = { "a":23,
              "b": 456,
              "c":222,
              "d":444}
```

```python
df1 = pd.DataFrame.from_dict(defaultMap, orient='index')
```

```python
df1.plot(kind="bar")
plt.savefig("./images/plot")
```

```python
defaultMap = { "JS - BFS":17,
              "CPP - BFS": 5.089,
              "JS - A*" : 15,
              "CPP - A*": 5.8}
df1 = pd.DataFrame.from_dict(defaultMap, orient='index')
df1.plot(kind="bar", legend=None)
plt.title("Default Assignment Map (64 x 64)")
plt.savefig("./images/defaultMap")
```

```python
bigCaves = { "JS - BFS":175,
              "CPP - BFS": 78,
              "JS - A*" : 192,
              "CPP - A*": 66.2}
df1 = pd.DataFrame.from_dict(bigCaves, orient='index')
df1.plot(kind="bar", legend=None)
plt.title("Large Caves (256 x 256)")
plt.savefig("./images/bigCaves")
```
